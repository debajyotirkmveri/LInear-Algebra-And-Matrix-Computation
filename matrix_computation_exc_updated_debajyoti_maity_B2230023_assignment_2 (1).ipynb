# Write a function to calculate the product of two matrices. Generate two random matrices $x \in \mathbf{R}^{m \times n}$ and $y \in \mathbf{R}^{n \times m}$ contains intergers between $[-l, l]$ and find the product of of $x$ and $y$ (your function should handle different order matrices). Check your result with the [numpy.matmul](https://numpy.org/doc/stable/reference/generated/numpy.matmul.html) or [numpy.dot](https://numpy.org/doc/stable/reference/generated/numpy.dot.html).

[10 points]




```python
m = 3
n = 5
l = 9
x = []
y = []

def matrix_multiplication(
    x: np.array,
    y: np.array,
)->np.array:
    """
    Dot product between two vectors
    
    Inputs:
        - x: 1st matrix
        - y: 2nd matrix
       
    Output:
        - val: multiplication x and y
    
    """
    
    val =  np.zeros((x.shape[0],y.shape[1]))
    ############################################################################
    #                             Your code will be here                       #
    #--------------------------------------------------------------------------#
    if (x.shape[1] == y.shape[0]):
        for i in range(x.shape[0]):
            for j in range(y.shape[1]):
                for k in range(y.shape[0]):
                    val[i][j] += x[i][k] * y[k][j]

    else:
        raise ValueError("Dimension Problem")
            
    #--------------------------------------------------------------------------#
    #                             End of your code                             #
    ############################################################################
    
    return val

# np.random.seed(42)# for reproduce the random numbers
A = np.random.randint(-l, l, size = (m,n))
B = np.random.randint(-l, l, size = (n,m))

print('Two random matrices are:')
print('A: \n{}\nand\nB: \n{}' .format(A, B))
print('-'*70)

val = matrix_multiplication(A, B)
print('The multiplication of A and B using my function is: \n{}' .format(val))
print('-'*70)

val1 = np.matmul(A, B)
print('The multiplication of A and B using numpy.matmul is: \n{}' .format(val1))
print('-'*70)

val1 = np.dot(A, B)
print('The multiplication of A and B using numpy.dot is: \n{}' .format(val1))
print('-'*70)
```

# Implement row and column sweep out and check your script with random example

```python
def sweep_out_row(
    A: np.array,
    ith_row: int,
    pivot_element: Tuple[int, int]
)->np.array:
    """
    Sweep out a i-th row based on the given pivot element
    
    Inputs:
        - A: Given matrix
        - ith_row: row to sweep out
        - pivot_element pivot element
        
    Output:
        - A: Resultant row sweep out matrix
    """
    
    ############################################################################
    #                             Your code will be here                       #
    #--------------------------------------------------------------------------#
    if pivot_element[0] != ith_row:
        raise ValueError ("pivot element position not in the given row ")
    elif A[pivot_element[0]-1,pivot_element[1]-1]==0:
        raise ValueError("pivot element is zero,sweeping cannot be done")
    else:
        A[:,pivot_element[1]-1]=A[:,pivot_element[1]-1]/A[pivot_element[0]-1,pivot_element[1]-1]
        for k in range (A.shape[1]):
            if k != pivot_element[1]-1 and A[pivot_element[0]-1,k]!=0:
                A[:,k]=A[:,k]-A[ith_row-1,k]*A[:,pivot_element[1]-1]    
        
    #--------------------------------------------------------------------------#
    #                             End of your code                             #
    ############################################################################
    
    return A
    
    
def sweep_out_col(
    A: np.array,
    ith_col: int,
    pivot_element: Tuple[int, int]
)->np.array:
    """
    Sweep out a i-th column based on the given pivot element
    
    Inputs:
        - A: Given matrix
        - ith_col: column to sweep out
        - pivot_element pivot element
        
    Output:
        - A: Resultant column sweep out matrix
    """
    
    ############################################################################
    #                             Your code will be here                       #

    #--------------------------------------------------------------------------#
    if pivot_element[1] != ith_col:
        raise ValueError ("pivot element position not in the given column ")
    elif A[pivot_element[0]-1,pivot_element[1]-1]==0:
        raise ValueError("pivot element is zero,sweeping cannot be done")
    else:
        A[pivot_element[0]-1,:]=A[pivot_element[0]-1,:]/A[pivot_element[0]-1,pivot_element[1]-1]
        for k in range (A.shape[0]):
            if k != pivot_element[0]-1 and A[k,pivot_element[1]-1]!= 0:
                A[k,:] = A[k,:]-A[k,ith_col-1]*A[pivot_element[0]-1,:]   
    
        
    #--------------------------------------------------------------------------#
    #                             End of your code                             #
    ############################################################################
    
    return A



np.random.seed(42)# for reproduce the random numbers
m = 3
n = 5
A =np.random.randint(10, 30, size = (m,n))

print('A random matrices A:\n{}' .format(A))
print('-'*70)
i = 1
j = 2
scalar = 3
E_ij = sweep_out_row(np.copy(A), i, (i, j))
print('{}-th row sweep out based on ({}, {}) as pivot on A: \n{}' .format(i, i, j, E_ij))
print('-'*70)

E_ij = sweep_out_col(np.copy(A), j, (i, j))
print('{}-th col sweep out based on ({}, {}) as pivot on A: \n{}' .format(j, i, j, E_ij))
print('-'*70)
```
